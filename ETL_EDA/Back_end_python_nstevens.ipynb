```python
# Import the dependency.
import pandas as pd
from pathlib import Path
```

```python
#importing lat and lng data from our North America dataset

lat_lng = Path("../Data/zip_to_lat_lon_North_America.csv")

lat_lng_df = pd.read_csv(lat_lng, low_memory=False)

lat_lng_df.tail()
```

```python
#Previewing the column names and counts
lat_lng_df.count()
```

```python
#get only the necessary column names
lat_lng_df_cleaned = lat_lng_df.loc[lat_lng_df['country code'] == 'US', ['postal code', 'latitude', 'longitude']]

lat_lng_df_cleaned.head()
```

```python
#confirming column names were dropped 
lat_lng_df_cleaned.count()
```

```python
#Renaming column name and data type to prepare the dataframe for merging 
lat_lng_df_cleaned = lat_lng_df_cleaned.rename(columns={'postal code': 'zip code'})
lat_lng_df_cleaned['zip code'] = lat_lng_df_cleaned['zip code'].astype(int)
lat_lng_df_cleaned.head()
```

```python
#Importing our second dataframe, which contains information on hospitals and providers
hospitals_data = Path("../Data/Medicare_Provider_Charge_Inpatient_DRGALL_FY2014.xlsx")

hospitals_df = pd.read_excel(hospitals_data, header = 5)

hospitals_df.head()
```

```python
#looking at the count and column names
hospitals_df.count()
```

```python
# see the data types
hospitals_df.info()
```

```python
#Renaming the zip code column to prepare it for a merge 
hospitals_df = hospitals_df.rename(columns={'Provider Zip Code': 'zip code'})

hospitals_df.head()
```

```python
#merging the two dataframes together based on the column 'zip code'
merged_data_df = pd.merge(hospitals_df, lat_lng_df_cleaned, on='zip code', how='inner')

merged_data_df.head()
```

```python
#Confirming whether we have to change the Dtype
merged_data_df.info()
```

```python
# see our provider ids
merged_data_df['Provider Id'].value_counts()
```

```python
# see extract DRG definition sort it by value counts then export to csv
drg_difinitions = merged_data_df['DRG Definition'].value_counts()
drg_difinitions.to_csv('../Data/drg_difinitions.csv')
```

```python
# extract the drg id from the drg finitition and place in it's own row
merged_data_df["DRG Id"] = merged_data_df["DRG Definition"].map(lambda x: x.split(' - ')[0])
merged_data_df.head()
```

```python
# extract the DRG information from database
drg_df = merged_data_df[["DRG Id", "DRG Definition"]].groupby('DRG Id').first().reset_index()
drg_df["Category"] = "Other"
drg_df.info()
drg_df.to_csv('../Data/drg.csv', index=False)
```

```python
# extract the provider information from the merged dataframe
provider_columns = ["Provider Id", "Provider Name", "Provider Street Address", 
                    "Provider City", "Provider State", "zip code", "latitude", "longitude"]

providers_df = merged_data_df[provider_columns].groupby('Provider Id').first().reset_index()
providers_df.info()
providers_df.to_csv('../Data/providers.csv', index=False)
```

```python
# extract the drg_records from the merged dataframe
drg_columns = ["DRG Id", "Provider Id", "Total Discharges", "Average Total Payments", 
                "Average Medicare Payments", "Average Covered Charges"]
drg_records_df = merged_data_df[drg_columns]
drg_records_df.info()
drg_records_df.to_csv('../Data/drg_records.csv', index=True)
```
